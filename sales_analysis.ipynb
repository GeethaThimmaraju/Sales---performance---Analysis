```python
# sales by region
region_sales = (df.groupby("Region")["Sales"].sum().reset_index().sort_values(by="Sales", ascending = True)) 
```

```python
region_sales
```

```python
#sales by category
category_sales = (df.groupby("Category")["Sales"].sum().reset_index().sort_values(by="Sales", ascending = True))
```

```python
category_sales
```

```python
# Sales by customer segment
CustomerSegment_sales = (df.groupby("Segment")["Sales"].sum().reset_index().sort_values(by="Sales", ascending = True))
```

```python
df["Order Date"].dtype
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d-%m-%Y")
```

```python
# how do sales change month by month
df["Order_year"] = df["Order Date"].dt.year
df["Order_month"] = df["Order Date"].dt.month
```

```python
# monthly sales aggregations
monthly_sales = (df.groupby(["Order_year", "Order_month"])["Sales"].sum().reset_index().sort_values(by=["Order_year", "Order_month"]))
```
